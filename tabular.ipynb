```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
t_df = pd.read_csv('tabular-playground-series-apr-2022/train.csv')
t_label_df = pd.read_csv('tabular-playground-series-apr-2022/train_labels.csv')
test_df = pd.read_csv('tabular-playground-series-apr-2022/test.csv')
total_df = pd.merge(t_df, t_label_df)
```

```python
total_df[['sequence', 'subject', 'step']].groupby(['sequence', 'subject']).count().reset_index()
```

```python
def createDataFrame(df):
    qdf = df.query('step == 0')
    return qdf
```

```python
logreg = LogisticRegression()
total_df = createDataFrame(total_df)
x = total_df.drop("state", axis=1)
y = total_df["state"]

X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.2, random_state=2)
```

```python
logreg.fit(X_train, Y_train)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
test_df = createDataFrame(test_df)
Y_pred = logreg.predict(test_df)
i = test_df['sequence'].to_numpy().reshape(1, len(Y_pred))
i = np.insert(i, 1, Y_pred, axis=0)
print(i.T.shape)
res = pd.DataFrame(i.T, columns = ['sequence','state'])
res.head()
```

```python
#indexをつけるとcolumns名無しのindexが1行目に入る
res.to_csv('result.csv', index=False)
```
